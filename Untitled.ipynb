```python
from pycaret.classification import *
import pandas as pd

df = pd.read_csv("bank-additional.csv", sep=';')

clf = setup(
    data=df,
    target='y',
    session_id=42,
    normalize=True,
    remove_multicollinearity=True,
    fold=5
)
```

```python
best_model = compare_models()
```

```python
tuned_model = tune_model(best_model)
```

```python
plot_model(tuned_model, plot='feature')
```

```python
rf = create_model('rf')        # Random Forest
plot_model(rf, plot='feature') # ✅ works
```

```python
ensemble_model = ensemble_model(tuned_model)
```

```python
evaluate_model(ensemble_model)
```

```python
final_model = finalize_model(ensemble_model)
```

```python
save_model(final_model, "loan_model")
```

```python
evaluate_model(final_model)
```
